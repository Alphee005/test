Comparons


*   ventes ~ tv + radio + journaux
*   ventes ~ tv + radio + journaux + tv^2
*   ventes ~ tv + radio + journaux + tv * radio


```python
df1['tv*radio']=df1.tv*df1.radio
variables=['tv','radio','journaux','tv2','TV*RADIO']
```

```python
regression= {
    'simple: y ~ tv + radio + journaux'  :     ['tv','radio','journaux'],
    'quadratique: y ~ tv + radio + journaux + tv2': ['tv','radio','journaux', 'tv2'],
    'terme croisée: y ~ tv + radio + journaux + tv*radio':['tv','radio','journaux', 'tv*radio']
}
```

```python
y=df1.ventes
```

```python
for titre, variables in regression.items():
  scaler=MinMaxScaler()
  data_array=scaler.fit_transform(df1[variables])
  data_scaler=pd.DataFrame(data_array,columns=variables)
  X=data_scaler[variables]
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
  reg=LinearRegression()
  reg.fit(X_train,y_train)
  y_pred=reg.predict(X_test)
  print(f"\n-- Regression {titre}")
  print(f"\tRMSE: {mean_squared_error(y_test, y_pred)}")
  print(f"\tMAPE: {mean_absolute_percentage_error(y_test, y_pred)}")
```
